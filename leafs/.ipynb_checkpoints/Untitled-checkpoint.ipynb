```python
def factorize(df):
    for column in df.columns:
        if (df[column].dtype != np.float64) and (df[column].dtype != np.int64):
            df[column] = df[column].factorize()[0]
```

```python
train = pd.read_csv('train.csv')
x_train = train.drop(['id', 'species'], axis=1)
le = LabelEncoder().fit(train['species'])
y_train = le.transform(train['species']) #classes to numbers
```

```python
contFeatureslist = []
for colName,x in x_train.iloc[1,:].iteritems():
    if(not str(x).isalpha()):
        contFeatureslist.append(colName)
```

```python
plt.figure(figsize=(13,9))
sns.boxplot(x_train[contFeatureslist])
plt.show()
```

```python
x_train_normalize = x_train[contFeatureslist]
x_train_normalize[contFeatureslist] = normalize(x_train[contFeatureslist])
```

```python
plt.figure(figsize=(13,9))
sns.boxplot(x_train_normalize[contFeatureslist])
plt.show()
```

```python
x_train_div = x_train
for feature in contFeatureslist:
    x_train_div[feature] = x_train[feature]/max(x_train[feature])
```
